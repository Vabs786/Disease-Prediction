### splitting dataset into traning and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

### Implementing KNearestNeighbour ALgo

```python
from sklearn.neighbors import KNeighborsClassifier 
no_neighbors = np.arange(1,10,2) 
print("KNN applied for ",no_neighbors," k values.") 
train_accuracy = np.empty(len(no_neighbors)) 
test_accuracy = np.empty(len(no_neighbors))

for i, k in enumerate(no_neighbors):
    knn=KNeighborsClassifier (n_neighbors=k) 
    knn.fit(X_train, np.ravel (y_train))

    train_accuracy[i] = knn.score(X_train, y_train) 
    
    test_accuracy[i] = knn.score(X_test, y_test)


plt.title('k-: Varying Number of Neighbors')
plt.plot(no_neighbors, test_accuracy, label='Testing Accuracy')
plt.plot(no_neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel("Accuracy") 
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train,np.ravel(y_train))
y_predknn=knn_model.predict(X_test)
y_predknn
```

```python
from sklearn.metrics import accuracy_score

print("Train Accuracy: ",round(accuracy_score(y_train, knn_model.predict (X_train))*100,2))
print("Test Accuracy: ",round(accuracy_score(y_test, y_predknn)* 100 ,2))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 42) 
dt.fit(X_train, y_train)
ypreddt=dt.predict(X_test)
print("Train Accuracy: ",round(accuracy_score(y_train, dt.predict (X_train))*100,2))
print("Test Accuracy: ",round(accuracy_score(y_test, ypreddt)*100,2))
```

```python
col= symp
ydt= dt.feature_importances_
fig, ax = plt.subplots()
width = 0.2
xdt = np.arange(len(ydt))
ax.barh(xdt, ydt, width, color="blue")
ax.set_yticks(xdt+width/10)
ax.set_yticklabels(col) 
plt.title("Feature importance in Decision tree classifier", fontsize=16)
plt.xlabel("relative importance", fontsize=14)
plt.ylabel("feature",fontsize=14)
plt.figure(figsize=(5,5))
fig.set_size_inches(15.5, 30.5 , forward=True)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=42)
logreg.fit(X_train,np.ravel(y_train))
ypredlogreg=logreg.predict(X_test)
print("Train Accuraccy:",round(accuracy_score(y_train,logreg.predict(X_train))*100,2))
print("Test Accuraccy:",round(accuracy_score(y_test, ypredlogreg)*100,2))
```
